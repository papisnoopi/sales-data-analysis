```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
top_teams = df.groupby('Team')['Goals For'].sum().sort_values(ascending=False).head(10)
print(top_teams)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top_teams.values, y=top_teams.index, palette='viridis')
plt.title('Top 10 Teams by Goals For - FIFA 2018')
plt.xlabel('Goals')
plt.ylabel('Team')
plt.show()
```

```python
df_plot = df[['Team', 'Win', 'Draw', 'Loss']].set_index('Team')
df_plot.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='Set2')
plt.title('Match Outcomes per Team - FIFA 2018')
plt.xlabel('Team')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Goal Difference', y='Team', data=df.sort_values(by='Goal Difference', ascending=False), palette='coolwarm')
plt.title('Goal Difference per Team - FIFA 2018')
plt.xlabel('Goal Difference')
plt.ylabel('Team')
plt.show()
```
